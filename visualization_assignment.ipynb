````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('college_completion.csv')
df.head()
````

````python
#check dimensions and variables
print("Data Dimensions:", df.shape)
print("\nVariables:\n", df.columns)

df.head()
````

````python
#cross tabulate control and level
control_level_ct = pd.crosstab(df['control'], df['level'])
print(control_level_ct)

#graph
control_level_ct.plot(kind='bar', stacked=True)
plt.title('Distribution of Institutions by Control and Level')
plt.xlabel('Control')
plt.ylabel('Count')
plt.show()
````

````python
import seaborn as sns

#histogram and kernel density plot 
plt.figure(figsize=(12, 6))
sns.histplot(df['grad_100_value'].dropna(), bins=30, kde=True)
plt.title('Histogram and Kernel Density of Graduation Rates')
plt.xlabel('Graduation Rate (100%)')
plt.show()

#boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['grad_100_value'])
plt.title('Boxplot of Graduation Rates')
plt.xlabel('Graduation Rate (100%)')
plt.show()

#stat description
print("\nStatistical Description:\n", df['grad_100_value'].describe())
````

````python
#grouped kernel density plot by control
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='grad_100_value', hue='control', common_norm=False)
plt.title('Kernel Density of Graduation Rates by Control')
plt.xlabel('Graduation Rate (100%)')
plt.show()

#grouped kernel density plot by level
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='grad_100_value', hue='level', common_norm=False)
plt.title('Kernel Density of Graduation Rates by Level')
plt.xlabel('Graduation Rate (100%)')
plt.show()

#grouped descriptions
print("\nGrouped by Control:\n", df.groupby('control')['grad_100_value'].describe())
print("\nGrouped by Level:\n", df.groupby('level')['grad_100_value'].describe())

#public, 2 year programs have highest graduation rates
````

````python
#new variable
df['levelXcontrol'] = df['level'] + ', ' + df['control']

#kernel density plot
plt.figure(figsize=(14, 7))
sns.kdeplot(data=df, x='grad_100_value', hue='levelXcontrol', common_norm=False)
plt.title('Kernel Density of Graduation Rates by Level and Control')
plt.xlabel('Graduation Rate (100%)')
plt.show()

#public 2 year colleges have best graduation rate
````

````python
#kernel density plot of aid value 
plt.figure(figsize=(12, 6))
sns.kdeplot(df['aid_value'].dropna())
plt.title('Kernel Density of Student Aid Value')
plt.xlabel('Average Aid Value')
plt.show()

#group by level and control
plt.figure(figsize=(14, 7))
sns.kdeplot(data=df, x='aid_value', hue='levelXcontrol', common_norm=False)
plt.title('Kernel Density of Aid Value by Level and Control')
plt.xlabel('Average Aid Value')
plt.show()

#most aid comes from 4-year private institutions, which makes sense

#stat descriptions
print("\nGrouped by Control:\n", df.groupby('control')['aid_value'].describe())
print("\nGrouped by Level:\n", df.groupby('level')['aid_value'].describe())
````

````python
#scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='aid_value', y='grad_100_value', data=df)
plt.title('Graduation Rate vs. Aid Value')
plt.xlabel('Average Aid Value')
plt.ylabel('Graduation Rate (100%)')
plt.show()

#higher the aid, typically more likely to graduate

#grouped by level
plt.figure(figsize=(10, 6))
sns.scatterplot(x='aid_value', y='grad_100_value', hue='level', data=df)
plt.title('Graduation Rate vs. Aid Value by Level')
plt.xlabel('Average Aid Value')
plt.ylabel('Graduation Rate (100%)')
plt.show()

# Grouped by Control
plt.figure(figsize=(10, 6))
sns.scatterplot(x='aid_value', y='grad_100_value', hue='control', data=df)
plt.title('Graduation Rate vs. Aid Value by Control')
plt.xlabel('Average Aid Value')
plt.ylabel('Graduation Rate (100%)')
plt.show()

#private 4-year institutions with aid help students graduate
````

$$
h(t) = \dfrac{f(t)}{1-F(t)}
$$

$$
H(t) = \int_0^t h(z) dz.
$$
