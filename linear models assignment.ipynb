```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("Q1_clean.csv")
df.head()
```

```python
#average price and review score by Neighbourhood
avg_stats = df.groupby("Neighbourhood ")[["Price", "Review Scores Rating"]].mean().sort_values(by="Price", ascending=False)

#most expensive borough
most_expensive_neighbourhood = avg_stats.idxmax()["Price"]
most_expensive_avg_price = avg_stats.max()["Price"]
print(f"the most expensive borough on average is {most_expensive_neighbourhood} with an average price of ${most_expensive_avg_price:.2f} per night.")
avg_stats

#kernel density plot of Price
plt.figure(figsize=(12,6))
sns.kdeplot(data=df, x="Price", hue="Neighbourhood ", common_norm=False)
plt.xlim(0, 1000) 
plt.title("Kernel Density Plot of Price by Neighbourhood")
plt.show()

#log transformation of Price (handling zero values)
df["Log Price"] = np.log1p(df["Price"])  

#kernel density plot of Log Price
plt.figure(figsize=(12,6))
sns.kdeplot(data=df, x="Log Price", hue="Neighbourhood ", common_norm=False)
plt.title("Kernel Density Plot of Log(Price) by Neighbourhood")
plt.show()
```

```python
#one-hot encode Neighbourhood (no intercept)
neighbourhood_dummies = pd.get_dummies(df["Neighbourhood "], drop_first=False)

#fit linear model without intercept
model_no_intercept = LinearRegression(fit_intercept=False)
X_no_intercept = neighbourhood_dummies
y = df["Price"]
model_no_intercept.fit(X_no_intercept, y)
coefficients_no_intercept = pd.Series(model_no_intercept.coef_, index=X_no_intercept.columns)

print("Coefficients in Regression Without Intercept:")
print(coefficients_no_intercept)
```

```python
#one-hot encode Neighbourhood 
neighbourhood_dummies_with_intercept = pd.get_dummies(df["Neighbourhood "], drop_first=True)

#fit linear model with intercept
model_with_intercept = LinearRegression(fit_intercept=True)
X_with_intercept = neighbourhood_dummies_with_intercept
model_with_intercept.fit(X_with_intercept, y)
intercept = model_with_intercept.intercept_
coefficients_with_intercept = pd.Series(model_with_intercept.coef_, index=X_with_intercept.columns)

print(f"Intercept: {intercept:.2f}")
print("Coefficients in Regression With Intercept:")
print(coefficients_with_intercept)
```

```python
#split data 80/20
X = df[["Review Scores Rating"]].join(neighbourhood_dummies_with_intercept)
y = df["Price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

#fit regression model
model = LinearRegression()
model.fit(X_train, y_train)

#predict and evaluate
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
review_score_coef = model.coef_[0]

print(f"R²: {r2:.4f}, RMSE: {rmse:.2f}")
print(f"Coefficient on Review Scores Rating: {review_score_coef:.2f}")
```

```python
#one-hot encode Property Type
property_dummies = pd.get_dummies(df["Property Type"], drop_first=True)

#combine features
X = df[["Review Scores Rating"]].join(neighbourhood_dummies_with_intercept).join(property_dummies)

#split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

#fit model
model_with_property = LinearRegression()
model_with_property.fit(X_train, y_train)

#predict and evaluate
y_pred_property = model_with_property.predict(X_test)
r2_property = r2_score(y_test, y_pred_property)
rmse_property = np.sqrt(mean_squared_error(y_test, y_pred_property))
review_score_coef_property = model_with_property.coef_[0]

#identify the most expensive property type
property_coefficients = pd.Series(model_with_property.coef_[1:], index=X.columns[1:])
most_expensive_property = property_coefficients.idxmax()
most_expensive_property_coef = property_coefficients.max()

print(f"R²: {r2_property:.4f}, RMSE: {rmse_property:.2f}")
print(f"Coefficient on Review Scores Rating: {review_score_coef_property:.2f}")
print(f"The most expensive property type to rent is {most_expensive_property}, with a coefficient of {most_expensive_property_coef:.2f}.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, OneHotEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

#convert categorical variables
df1['Make_Year'] = df1['Make_Year'].astype(int)

#handle outliers
df1['Price_log'] = np.log1p(df1['Price'])
df1['Mileage_Run_log'] = np.log1p(df1['Mileage_Run'])
```

```python
#summarize price variable
price_summary = df1['Price'].describe()

#kernel density plot for price
plt.figure(figsize=(8, 5))
sns.kdeplot(df1['Price'], fill=True, bw_adjust=0.5)
plt.xlabel("Price")
plt.ylabel("Density")
plt.title("Kernel Density Plot of Price")
plt.show()

#summarize prices by brand
brand_price_summary = df1.groupby("Make")["Price"].describe()

#grouped kernel density plot by Make
plt.figure(figsize=(12, 6))
for make in df1["Make"].unique():
    sns.kdeplot(df1[df1["Make"] == make]["Price"], label=make, fill=True, alpha=0.3)
plt.xlabel("Price")
plt.ylabel("Density")
plt.title("Kernel Density Plot of Price by Car Brand")
plt.legend()
plt.show()
```

```python
#split data into training and test sets 
train, test = train_test_split(df1, test_size=0.2, random_state=42)
```

```python
#define numeric features
numeric_features = ["Make_Year", "Mileage_Run_log", "Seating_Capacity"]

#train model using only numeric variables
X_train_num = train[numeric_features]
X_test_num = test[numeric_features]
y_train = train["Price"]
y_test = test["Price"]

#standardize numeric features
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)

#fit regression model
model_num = LinearRegression()
model_num.fit(X_train_num_scaled, y_train)
y_train_pred = model_num.predict(X_train_num_scaled)
y_test_pred = model_num.predict(X_test_num_scaled)
r2_train_num = r2_score(y_train, y_train_pred)
r2_test_num = r2_score(y_test, y_test_pred)
rmse_train_num = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test_num = np.sqrt(mean_squared_error(y_test, y_test_pred))

#one-hot encode categorical variables
X_train_cat = pd.get_dummies(train.drop(columns=["Price"]), drop_first=True)
X_test_cat = pd.get_dummies(test.drop(columns=["Price"]), drop_first=True)
X_test_cat = X_test_cat.reindex(columns=X_train_cat.columns, fill_value=0)

#train categorical model
model_cat = LinearRegression()
model_cat.fit(X_train_cat, y_train)
y_test_pred_cat = model_cat.predict(X_test_cat)
r2_test_cat = r2_score(y_test, y_test_pred_cat)
rmse_test_cat = np.sqrt(mean_squared_error(y_test, y_test_pred_cat))

#combined model
X_train_combined = pd.concat([pd.DataFrame(X_train_num_scaled, columns=numeric_features), X_train_cat.reset_index(drop=True)], axis=1)
X_test_combined = pd.concat([pd.DataFrame(X_test_num_scaled, columns=numeric_features), X_test_cat.reset_index(drop=True)], axis=1)
model_combined = LinearRegression()
model_combined.fit(X_train_combined, y_train)
y_test_pred_combined = model_combined.predict(X_test_combined)
r2_test_combined = r2_score(y_test, y_test_pred_combined)
rmse_test_combined = np.sqrt(mean_squared_error(y_test, y_test_pred_combined))
```

```python
#polynomial feature expansion
poly_degrees = range(1, 6)
r2_test_poly = []
rmse_test_poly = []
for degree in poly_degrees:
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_train_poly = poly.fit_transform(X_train_num_scaled)
    X_test_poly = poly.transform(X_test_num_scaled)
    model_poly = LinearRegression()
    model_poly.fit(X_train_poly, y_train)
    y_test_pred_poly = model_poly.predict(X_test_poly)
    r2_test_poly.append(r2_score(y_test, y_test_pred_poly))
    rmse_test_poly.append(np.sqrt(mean_squared_error(y_test, y_test_pred_poly)))

#r^2 goes negative
negative_r2_degree = next((d for d, r2 in zip(poly_degrees, r2_test_poly) if r2 < 0), None)
```

```python
#best model
best_degree = poly_degrees[np.argmax(r2_test_poly)]
best_r2 = max(r2_test_poly)
best_rmse = min(rmse_test_poly)
poly = PolynomialFeatures(degree=best_degree, include_bias=False)
X_train_poly_best = poly.fit_transform(X_train_num_scaled)
X_test_poly_best = poly.transform(X_test_num_scaled)
model_poly_best = LinearRegression()
model_poly_best.fit(X_train_poly_best, y_train)
y_test_pred_best = model_poly_best.predict(X_test_poly_best)
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_test_pred_best, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='dashed')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Predicted vs. Actual Price for Best Model")
plt.show()

residuals = y_test - y_test_pred_best
plt.figure(figsize=(8, 5))
sns.kdeplot(residuals, fill=True, bw_adjust=0.5)
plt.axvline(x=0, color='red', linestyle='dashed')
plt.xlabel("Residuals")
plt.ylabel("Density")
plt.title("Kernel Density Plot of Residuals")
plt.show()

#model evaluations
r2_train_num, r2_test_num, rmse_train_num, rmse_test_num, r2_test_cat, rmse_test_cat, r2_test_combined, rmse_test_combined, best_r2, best_rmse, negative_r2_degree
```

**Q10.** Let's look at a cousin of Linear Regression, called **kernel regression** or **local constant least squares** or **Nadaraya-Watson Estimator**.

We derived the OLS estimator for single linear regression by minimizing
$$
SSE(b_0, b_1) = \frac{1}{N}\sum_{i=1}^N (y_i - b_0 - b_1 x_i)^2
$$
with solution
$$
\hat{b}_0 = \bar{y} - \hat{b}_1 \bar{x}, \quad \hat{b}_1 = \dfrac{\frac{1}{N} \sum_{i=1}^N (x_i-\bar{x})(y_i - \bar{y})}{s_x^2}.
$$

When you step back and think about it, this is a bit weird: The algorithm is computing sample means, variances, and covariances, and using those to create a predictive model. The data themselves arguably vanish from the solution. This is elegant, this is strange.

Instead, let $k(z)$ be a kernel function, such as the Gaussian
$$
k(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}
$$
or uniform
$$
k(z) = \begin{cases}
1/2, & |z| \le \frac{1}{2}\\
0, & \text{otherwise.}
\end{cases}
$$
We'll instead minimize, or each predictor value $x$,
$$
